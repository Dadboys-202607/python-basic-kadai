```python
#ランダムな整数を利用するために、randomモジュールをインポート
import random

# 変数varに1～40までのランダムな整数を代入する
var = random.randint(1, 40)

# 変数varの値を出力して確認
print(var)

# 変数varの値が3の倍数であれば「Fizz」という文字列を出力する
# 変数varの値が5の倍数であれば「Buzz」という文字列を出力する
# 変数varの値が3と5両方の倍数であれば「FizzBuzz」という文字列を出力する
if var % 3 == 0 and var % 5 == 0:
    print("FizzBuzz")
elif var % 3 == 0:
    print("Fizz")
elif var % 5 == 0:
    print("Buzz")
else:
    print(var)
```
