```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        print(self.name, self.age)
        if self.age >= 20:
            print("大人です")
        else:
            print("大人ではありません")
        print()


Shain1 = Human("Tanaka", 5)
Shain2 = Human("Suzuki", 25)
Shain3 = Human("Yamada", 20)
Shain4 = Human("Sato", 15)
Shain5 = Human("Watanabe", 30)
Shain6 = Human("Ikeda", 5)

Shain_list = [Shain1, Shain2, Shain3, Shain4, Shain5, Shain6]

i = 0
while i < len(Shain_list):
    Shain_list[i].check_adult()
    i += 1
```
